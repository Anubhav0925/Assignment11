Q1. Explain why we have to use the Exception class while creating a Custom Exception. in python

Answere :- In Python, exceptions are used to indicate that an error or unexpected event has occurred during the execution of a program. Python provides a set of built-in exceptions that can be raised by the interpreter or the programmer. However, sometimes the built-in exceptions are not sufficient to convey the specific nature of an error or problem in the program . When we encounter such a scenario where the built-in exceptions are not enough, we can create our own exceptions, which are known as custom exceptions. These custom exceptions are created by creating a new class that inherits from the Exception class. The reason for using the Exception class as the base class for custom exceptions is that it provides a standard interface for handling exceptions. By inheriting from the Exception class, our custom exception class will have access to all the methods and attributes of the Exception class, which can be used to handle the exception in a consistent and standardized way.For example, when we catch an exception, we can use the 'except' clause followed by the type of the exception, and the code in the 'except' block will only execute if the exception is of that type or a subclass of it. By inheriting from the Exception class, our custom exception class can be caught and handled in the same way as any other exception. In summary, we use the Exception class while creating a custom exception in Python because it provides a standard interface for handling exceptions and allows our custom exception class to be caught and handled in the same way as any other exception.

Q2. Write a python program to print Python Exception Hierarchy.

```python
# Answere :-
# print Python Exception Hierarchy
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + str(exception_class))
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 2)

# start from the base class Exception
print_exception_hierarchy(Exception)
```

This program defines a function called 'print_exception_hierarchy' that takes an exception class and an optional 'indent' parameter. It then prints the name of the exception class, indented by the number of spaces specified by the 'indent' parameter. It then recursively calls itself on all the subclasses of the exception class, passing in the subclass and an increased 'indent' value.The program then starts the recursive process by calling 'print_exception_hierarchy' on the base class 'Exception'. When you run this program, it will output the Python Exception Hierarchy in a tree-like structure, with each exception class indented to show its position in the hierarchy.

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Answere :- The 'ArithmeticError' class is a built-in exception class in Python that is raised when an arithmetic operation fails to execute properly. It is a base class for all errors that occur during arithmetic calculations. Some common errors that are defined in the 'ArithmeticError' class are 'ZeroDivisionError', OverflowError, and 'FloatingPointError'. Here are explanations of two of them with examples:

```python
# 1. ZeroDivisionError: This error is raised when attempting to divide by zero. For example:
# division by zero
try:
    x = 10/0
except ZeroDivisionError as error:
    print(f"Caught an exception: {error}")
```

2. OverflowError: This error is raised when the result of an arithmetic operation is too large to be represented by the available memory. For example:

```python
# integer overflow
import sys
try:
    x = sys.maxsize + 1
except OverflowError as error:
    print(f"Caught an exception:{error}")
```

In the first example, we attempt to divide the number 10 by 0, which is not a valid arithmetic operation, and thus a 'ZeroDivisionError' is raised. In the second example, we attempt to add 1 to the maximum integer value that can be represented by the system, which results in an overflow and a 'OverflowError' is raised.
These are just two examples of errors that are defined in the 'ArithmeticError' class. There are several other arithmetic-related errors that can be raised during program execution, depending on the specific scenario in which the error occurs.

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Answere :- The 'LookupError' class is a built-in exception class in Python that is raised when an index or key used to access an object is invalid. It is the base class for all errors that occur during indexing or key lookup.In general, the 'LookupError' class is used to handle situations where we are trying to access an object using an index or key that does not exist. This can happen with different types of objects like lists, dictionaries, tuples, and so on. Two common subclasses of 'LookupError' are 'KeyError' and 'IndexError'. Here are explanations of these two exceptions with examples:

```python
#1. 'KeyError': This error is raised when we try to access a dictionary using 
# a key that does not exist in the #dictionary. For example:

# KeyError example
my_dict = {"a": 1, "b": 2, "c": 3}
try:
    x = my_dict["d"]
except KeyError as error:
    print(f"Caught an exception: {error}")
```

In this example, we try to access the value associated with the key "d" in the 'my_dict' dictionary, which does not exist. Thus, a 'KeyError' is raised.

```python
#2. IndexError: This error is raised when we try to access an item in a list or tuple 
#    using an index that is out of range. For example:

# IndexError example
my_list = [1, 2, 3]
try:
    x = my_list[3]
except IndexError as error:
    print(f"Caught an exception: {error}")
```

In this example, we try to access the item at index 3 in the 'my_list' list, which is out of range since the list only contains three items. Thus, an 'IndexError' is raised.  These are just two examples of errors that are subclasses of the 'LookupError' class. In general, the 'LookupError' class is used to handle situations where we are trying to access an object using an index or key that does not exist.

Q5. Explain ImportError. What is ModuleNotFoundError?

Answere :- In Python, modules are reusable pieces of code that can be imported into other Python scripts or modules to add specific functionalities or features. When a Python script tries to import a module and fails to do so, it raises an error called ImportError. ImportError occurs when Python cannot locate a module that a script or module is trying to import. It can occur due to various reasons such as incorrect installation of the module, incorrect path to the module, or the module itself might not exist. For example, if a script tries to import a module called "numpy" but the module is not installed in the system, an ImportError will be raised. On the other hand, ModuleNotFoundError is a more specific error that is raised when Python cannot find the module that a script or module is trying to import. It is a subclass of ImportError and was introduced in Python 3.6 to provide more detailed information about the error. ModuleNotFoundError occurs when the specified module does not exist, or the module name is misspelled, or the module is not installed in the system. It can also occur when the module is located in a different directory or when the script is executed in a virtual environment that does not have the module installed. For example, if a script tries to import a module called "numpy" but misspells the name as "nump", a ModuleNotFoundError will be raised.

Q6. List down some best practices for exception handling in python.

Answere :- Exception handling is an essential aspect of programming in Python as it allows for the graceful handling of errors and unexpected situations that might occur during program execution. Here are some best practices for exception handling in Python:
1. Use specific exception types: It is a good practice to use specific exception types instead of catching general exceptions like the "Exception" class. This makes the code more readable and helps to handle specific errors more effectively.
2. Use try-except blocks: The try-except block is a widely used construct in Python for handling exceptions. The code inside the try block is executed, and if any exception is raised, the code inside the corresponding except block is executed. It is recommended to use multiple except blocks for handling different types of exceptions.
3. Avoid bare except clauses: A bare except clause is a catch-all exception handler that catches all exceptions, including those that were not intended to be caught. This can make it difficult to diagnose and fix errors. Instead, use specific exception handlers that handle only the exceptions that are expected to occur.
4. Use finally block: The finally block is executed regardless of whether an exception is raised or not. It is used to perform cleanup actions such as closing files, database connections, or releasing resources.
5. Use the "with" statement: The "with" statement is used to ensure that resources are properly released, even if an exception occurs. It is recommended to use the "with" statement for handling file I/O, database connections, and other resources that need to be released after use.
6. Log exceptions: Logging exceptions is a good practice as it helps to debug the code and identify the root cause of the error. Python's built-in logging module provides an easy way to log exceptions and other information.
7. Raise exceptions: It is a good practice to raise exceptions when an error occurs, rather than ignoring it or returning an error code. This helps to identify the root cause of the error and handle it effectively.
